```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
load_data = os.path.join('Data', 'parkinsons.csv')
```

```python
#read data
read_data = pd.read_csv(load_data)
read_data
```

```python
read_data = read_data.drop('name', 1)
read_data
```

```python
#corr
corr_features = read_data.corr(method='pearson')
corr_features
```

```python
corr_features1 = read_data.corr(method='kendall')
corr_features1
```

```python
#plot the corr data (pearson)
plot1 = sb.heatmap(corr_features)
```

```python
#plot corr data (pearson)
plt.figure(figsize=(18, 8))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(read_data.corr(method= 'pearson'), dtype=np.bool))
heatmap = sb.heatmap(read_data.corr(method= 'pearson'), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation of Features', fontdict={'fontsize':20}, pad=16);
```
